```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.integrate as integrate
from scipy.signal import find_peaks
```

```python
# 50Ga
n = 10000
t = 5
t_range = np.linspace(0, n-1, n) * t

#radii
# mass variation (alt50k)
radiuses_m5 = np.loadtxt('variable_mass/mass_e5/Radiuses.txt')
radiuses_m6 = np.loadtxt('variable_mass/mass_e6/Radiuses.txt')
radiuses_m7 = np.loadtxt('variable_mass/mass_e7/Radiuses.txt')
# altitude variation (m6)
radiuses_alt100k = np.loadtxt('variable_altitude/alt_100k/Radiuses.txt')
radiuses_alt50k = np.loadtxt('variable_altitude/alt_50k/Radiuses.txt')
radiuses_alt10k = np.loadtxt('variable_altitude/alt_10k/Radiuses.txt')
radiuses_alt5k = np.loadtxt('variable_altitude/alt_5k/Radiuses.txt')
radiuses_alt3k = np.loadtxt('variable_altitude/alt_3k/Radiuses.txt')

#friction
# mass variation (alt50k)
friction_m5 = np.loadtxt('variable_mass/mass_e5/Friction.txt')
friction_m6 = np.loadtxt('variable_mass/mass_e6/Friction.txt')
friction_m7 = np.loadtxt('variable_mass/mass_e7/Friction.txt')
# altitude variation (m6)
friction_alt3k = np.loadtxt('variable_altitude/alt_3k/Friction.txt')
friction_alt5k = np.loadtxt('variable_altitude/alt_5k/Friction.txt')
friction_alt10k = np.loadtxt('variable_altitude/alt_10k/Friction.txt')
friction_alt50k = np.loadtxt('variable_altitude/alt_50k/Friction.txt')
friction_alt100k = np.loadtxt('variable_altitude/alt_100k/Friction.txt')
```

```python
plt.figure(figsize=(10,10))
plt.plot(t_range, radiuses_m5/np.amax(radiuses_m5), label='M = 10^5 Msol')
plt.plot(t_range, radiuses_m6/np.amax(radiuses_m6), label='M = 10^6 Msol')
plt.plot(t_range, radiuses_m7/np.amax(radiuses_m7), label='M = 10^7 Msol')
plt.legend(fontsize=15)
```

```python
plt.figure(figsize=(10,10))
#plt.plot(t_range, radiuses_alt100k/np.amax(radiuses_alt100k), label='R_0 = 100kpc')
plt.plot(t_range, radiuses_alt50k /np.amax(radiuses_alt50k ), label='R_0 = 50kpc')
plt.plot(t_range, radiuses_alt10k /np.amax(radiuses_alt10k ), label='R_0 = 10kpc')
plt.plot(t_range, radiuses_alt5k /np.amax(radiuses_alt5k ), label='R_0 = 5kpc')
#plt.plot(t_range, radiuses_alt3k /np.amax(radiuses_alt3k), label='R_0 = 3kpc')
plt.legend(fontsize=15)
plt.title('Orbital Decay for Different Starting Altitudes', fontsize=20)
plt.ylabel('R/R_0', fontsize=20)
plt.xlabel('t /Ma', fontsize=20)
plt.savefig('orbitaldecay_altitudes.png')
```

```python
plt.figure(figsize=(10,10))
plt.plot(t_range, np.log10(friction_m5), label='M = 10^5 Msol')
plt.plot(t_range, np.log10(friction_m6), label='M = 10^6 Msol')
plt.plot(t_range, np.log10(friction_m7), label='M = 10^7 Msol')
plt.legend(fontsize=15)
```

```python
plt.figure(figsize=(10,10))
#plt.plot(t_range, np.log10(friction_alt100k), label='R_0 = 100kpc')
plt.plot(t_range, np.log10(friction_alt50k ), label='R_0 = 50kpc')
plt.plot(t_range, np.log10(friction_alt10k ), label='R_0 = 10kpc')
plt.plot(t_range, np.log10(friction_alt5k ), label='R_0 = 5kpc')
plt.legend(fontsize=15)
plt.title('Deceleration from Friction over time', fontsize=20)
plt.xlabel('t /Ma', fontsize=20)
plt.ylabel('log(a/ (pc/Ma^2))', fontsize=20)
plt.savefig('friction_altitudes.png')
```
